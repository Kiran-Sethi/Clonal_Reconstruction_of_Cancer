```python
##for pedline
data['cellName_splitted']=data['cellName'].apply(lambda x: x.split("-"))
data['cellType_splitted']=data['cellType'].apply(lambda x:x.split("_"))
```

```python
def separator(row):
    if "-" in row:
     lst=row.split("-")
    elif "_" in row:
     lst=row.split("_")
    return lst

##for melanoma
## for glioblastoma also
data['cellName_splitted']=data['cellName'].apply(separator)

data['cellType_splitted']=data['cellType'].apply(lambda x:x.split("_"))
```

```python
data['patient']=data['cellName_splitted'].apply(lambda x:x[0])
```

```python
lst={}
for index,row in data.iterrows():
  ele=row['cellName_splitted'][0]
  if ele not in lst:
    lst[ele]=1
  else:
    lst[ele]+=1
```

```python
oligo_path=os.path.join(base_path,"oligo")
lst=[df.get_group(x) for x in df.groups]
for patient_df in lst:
  for ele in set(patient_df['patient']):
    name=ele
  patient_df.drop(columns=['cellName_splitted','cellType_splitted','patient'],inplace=True)
  patient_df.to_csv(oligo_path+name+".txt",index=False,header=False,sep='\t')
  print(oligo_path+name+".txt")
```

```python
melanoma_path=os.path.join(base_path,"melanoma")
lst=[df.get_group(x) for x in df.groups]
for patient_df in lst:
  for ele in set(patient_df['patient']):
    name=ele
  patient_df.drop(columns=['cellName_splitted','cellType_splitted','patient'],inplace=True)
  patient_df.to_csv(os.path.join(melanoma_path,"reference"+name+".txt"),index=False,header=False,sep='\t')
```

```python
glioblastoma_path=os.path.join(base_path,"glioblastoma")
lst=[df.get_group(x) for x in df.groups]
for patient_df in lst:
  for ele in set(patient_df['patient']):
    name=ele
  patient_df.drop(columns=['cellName_splitted','cellType_splitted','patient'],inplace=True)
  if name=="GTEX":
     patient_df.to_csv(os.path.join(glioblastoma_path,"reference"+name+".txt"),index=False,header=False,sep='\t')
     break

patient_df.head()
refernce=patient_df.copy()
```

```python
glioblastoma_path=os.path.join(base_path,"glioblastoma")
lst=[df.get_group(x) for x in df.groups]
for patient_df in lst:
  for ele in set(patient_df['patient']):
    name=ele
  patient_df.drop(columns=['cellName_splitted','cellType_splitted','patient'],inplace=True)
  if name!="GTEX":
     patient_df=pd.concat([patient_df,refernce])
     patient_df.to_csv(os.path.join(glioblastoma_path,"glioblastoma_"+name+".txt"),index=False,header=False,sep='\t')
```

```python
patient_annot=[]
lst=[df.get_group(x) for x in df.groups]

for patient_df in lst:
  for ele in set(patient_df['patient']):
    name=ele
    patient_annot.append("melanoma"+name+".txt")
```
